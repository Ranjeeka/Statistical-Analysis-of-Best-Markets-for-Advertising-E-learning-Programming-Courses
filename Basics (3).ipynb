# Identifying the right market to targer for advertising courses from an e-learning company #

Our task is to identify the right target market for the e-learning company to advertise in. We are looking to to get two best markets.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
from numpy import mean, std
import matplotlib.pyplot as plt
data = pd.read_csv('2017-fCC-New-Coders-Survey-Data.csv', low_memory=False)
data.head(5)
```

The survey we are using is from [freecodecamp's 2017 New coder survey](https://www.freecodecamp.org/news/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969). [Available in this Github repositary](https://github.com/freeCodeCamp/2017-new-coder-survey)

### Is the sample representative of our population of interest? ###

```python
Fre_table = data['JobRoleInterest'].value_counts(normalize = True).reset_index()
```

```python
#split job interests to count mobile and web development entries
f_dic = {}
data['JobRoleInterest'] = data['JobRoleInterest'].str.split(',')
data['JobRoleInterest'] = data['JobRoleInterest'].fillna(0)
```

```python
#counting and adding key value pairs to dictionary
for n in data['JobRoleInterest']:
    if n!= 0:
        for val in n:
            if val in f_dic:
                f_dic[val] +=1
            else:
                f_dic[val] = 1
```

```python
#combining multiple similar entry counts for web and mobile developement.
sums = [0,0,0]
for key in f_dic:
    if 'Web' in key:
        sums[1] += f_dic[key]
    elif 'Mobile' in key:
        sums[0] += f_dic[key]
        
```

```python
total = sum(f_dic.values())
sums[2] = total - (sums[0] +sums[1])
```

```python
%matplotlib inline
mylabels = ['mobile', 'web', 'others']
fig = plt.figure(figsize =(10, 7))
plt.pie(sums, labels=mylabels)
plt.show()
```

The above graph shows us that collective interest in mobile and web development is over 50% of the total usable responses. The unusable responses with nan values have been excluded from the final count.

Many people are interested in more than one subject but the sample is representative beacuse having varied interest doesn't mean they'd not be interested in the courses offered by the client. The graph above proves that there is a market for web development.

### Eliminating Outliers ###
We eliminated outliers using their z-scores which is (value-mean)/std. Std is standard deviation of that array.
Where z-score is 0, it means that the value is exactly equal to mean because std of mean =0.
Where z-score is 1, it means that the value is 1 standard deviation away from mean and so forth.
We consider the values over z-score 3 to be outliers.

```python
#dropping outliers

def z(arr):
    col=[]
    for x in arr:
        col.append((x - mean(arr)) / std(arr,ddof=0))
    return col

z_data = data_cl.pivot(columns = 'CountryLive', values = 'MonthlySpent').reset_index()

z_data['zcan'] = z(z_data['Canada'])
z_data['zin'] = z(z_data['India'])
z_data['zuk'] = z(z_data['United Kingdom'])
z_data['zus'] = z(z_data['United States of America'])
z_can = z_data[['Canada','zcan']].copy()
z_uk = z_data[['United Kingdom','zuk']].copy()
z_in = z_data[['India','zin']].copy()
z_us= z_data[['United States of America','zus']].copy()
```

```python
z_can.dropna(inplace=True)
z_can = z_can[abs(z_can['zcan']) < 3].copy()
z_uk.dropna(inplace=True)
z_uk = z_uk[abs(z_uk['zuk']) < 3].copy()
z_in.dropna(inplace=True)
z_in = z_in[abs(z_in['zin']) < 3].copy()
z_us.dropna(inplace=True)
z_us = z_us[abs(z_us['zus']) < 3].copy()

print(z_can.mean(), z_uk.mean(), z_in.mean(), z_us.mean())


```

### Second market of focus advertisements in? ###
